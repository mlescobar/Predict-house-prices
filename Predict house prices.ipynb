 # Predicting houses 
 
The following project aims to predict the price of a house in Boston based on the number of rooms it has. This project uses a simple linear regression method.

```python
import numpy as np
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

#Preparing the data
boston = datasets.load_boston()
print(boston)
print()

#Verify the dataset information
print('Dataset info:')
print(boston.keys())
print()

#Dataset features
print('Dataset features:')
print(boston.DESCR)

#Data size
print('Data size:')
print(boston.data.shape)
print()

#Column names
print('Column names:')
print(boston.feature_names)

#We select only column 5 of the dataset because there is the number of rooms of the house
X = boston.data[:, np.newaxis, 5]

#Data on target
y = boston.target


plt.scatter(X,y)
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()

#We split de dataset in training data and test data (20% test data)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

#Algorithm
lr = linear_model.LinearRegression()

#Model training
lr.fit(X_train,y_train)

#Model prediction
Y_pred = lr.predict(X_test)

#Data visualization (and the model)
plt.scatter(X_test,y_test)
plt.plot(X_test, Y_pred, color='red',linewidth=3)
plt.title('Simple linear regression')
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()

print()
print('SIMPLE LINEAR REGRESSION MODELS DATA')
print()
print('Slope value or coefficient "a:')
print(lr.coef_)
print()
print('Slope value or coefficient "b:')
print(lr.intercept_)
print()
print('Model equation:')
print('y = ', lr.coef_, 'x', lr.intercept_)
print()
print('Model precision')
print(lr.score(X_train,y_train))
print()
```
