# Predicting houses 
 
The following project aims to predict the price of a house in Boston based on the number of rooms it has. This project uses a support vector machine method.

```python
import numpy as np
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR

#Preparing the data
boston = datasets.load_boston()

"""
print(boston)
print()

#VVerify the datasat information
print('Dataset info:')
print(boston.keys())
print()

#Dataset features
print('Dataset features:')
print(boston.DESCR)

#Data size
print('Data size:')
print(boston.data.shape)
print()

#Column names
print('Column names:')
print(boston.feature_names)

"""

#We split de datasat in training data and test data (20% test data)
X = boston.data[:, np.newaxis, 5]

#Data on target
y = boston.target

plt.scatter(X,y)
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()

#We split de datasat in training data and test data (20% test data)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

#Algorithm
svr = SVR(kernel='linear',C=1.0,epsilon=0.2)

#Model training
svr.fit(X_train,y_train)

#Model prediction
Y_pred = svr.predict(X_test)

#Data visualization (and the model)
plt.scatter(X_test,y_test)
plt.plot(X_test, Y_pred, color='red',linewidth=3)
plt.show()

print()
print('SUPPORT VECTOR MACHINE DATA')
print()
print('Model precision')
print(svr.score(X_train,y_train))
print()

```
