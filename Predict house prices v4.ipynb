 # Predicting houses 
 
The following project aims to predict the price of a house in Boston based on the number of rooms it has. This project uses decision tree regression method.

```python
import numpy as np
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

#Preparing the data
boston = datasets.load_boston()

"""
print(boston)
print()

#VVerify the datasat information
print('Dataset info:')
print(boston.keys())
print()

#Dataset features
print('Dataset features:')
print(boston.DESCR)

#Data size
print('Data size:')
print(boston.data.shape)
print()

#Column names
print('Column names:')
print(boston.feature_names)

"""

#We split de datasat in training data and test data (20% test data)
X = boston.data[:, np.newaxis, 5]

#Data on target
y = boston.target

plt.scatter(X,y)
plt.xlabel('Número de habitaciones')
plt.ylabel('Valor medio')
plt.show()

#We split de datasat in training data and test data (20% test data)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

#Algorithm
adr = DecisionTreeRegressor(max_depth = 5)

#Model training
adr.fit(X_train,y_train)

#Model prediction
Y_pred = adr.predict(X_test)

#Data visualization (and the model)
X_grid = np.arange(min(X_test),max(X_test),0.1)
X_grid = X_grid.reshape((len(X_grid),1))
plt.scatter(X_test,y_test)
plt.plot(X_grid, adr.predict(X_grid), color='red',linewidth=3)
plt.show()

print()
print('DECISION TREE REGRESSION DATA')
print()
print('Model precision')
print(adr.score(X_train,y_train))
print()

```
