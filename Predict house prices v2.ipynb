# Predicting houses 
 
The following project aims to predict the price of a house in Boston based on the number of rooms it has. This project uses a polynomial regression method.

```python
import numpy as np
from sklearn import datasets,linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

#Preparing the data
boston = datasets.load_boston()

#We select only column 5 of the dataset because there is the number of rooms of the house
X = boston.data[:, np.newaxis, 5]

#Data on target
y = boston.target

plt.scatter(X,y)
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()

#We split de datasat in training data and test data (20% test data)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

#Polynom degree
poli_reg = PolynomialFeatures(degree = 2)

#Transforming features to a higher degree
X_train_t = poli_reg.fit_transform(X_train)
X_test_t = poli_reg.fit_transform(X_test)

#Algorithm
lr = linear_model.LinearRegression()
#Model training
lr.fit(X_train,y_train)

#Model prediction
Y_pred = lr.predict(X_test)

#Data visualization (and the model)
plt.scatter(X_test,y_test)
plt.plot(X_test, Y_pred, color='red',linewidth=3)
plt.title('Polynomial regression')
plt.xlabel('Number of rooms')
plt.ylabel('Mean value')
plt.show()

print()
print('POLYNOMIAL REGRESSION MODELS DATA')
print()
print('Slope value or coefficient "a:')
print(lr.coef_)
print()
print('Slope value or coefficient "b:')
print(lr.intercept_)
print()
print('Model equation:')
print('y = ', lr.coef_, 'x', lr.intercept_)
print()
print('Model precision')
print(lr.score(X_train,y_train))
print()
```
